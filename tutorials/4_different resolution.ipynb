# Combinging different resolution

---

Say that we have a high resolution raster, and we want to combine it with a low resoltion vector classifier. We don't want to oversample the vector: 

```python
import sys
sys.path.append('..')

from agrid.grid import Grid
from agrid.acc import download

world = Grid(crs=4326, res = [0.5, 0.5], left = -180, up=90, right = 180, down = -90)
```

```python
world.ds['RGB_RASTER'] = (('Y', 'X', 'RGB'), 
                world.read_raster('../../data/ne/HYP_50M_SR_W/HYP_50M_SR_W.tif', sub_sampling=5) )
world.map_grid('RGB_RASTER')
```
